```python
from weka.core.converters import Loader
data_dir = "/Users/rmontanana/Code/discretizbench/datasets/"
loader = Loader(classname="weka.core.converters.ArffLoader")
data = loader.load_file(data_dir + "iris.arff")
data.class_is_last()
```

```python
print(data)
```

```python
from weka.classifiers import Classifier
cls = Classifier(classname="weka.classifiers.trees.J48", options=["-C", "0.3"])
```

```python
print(cls.options)
```

```python
print(cls.to_commandline())
```

```python
from weka.classifiers import Classifier
cls = Classifier(classname="weka.classifiers.bayes.BayesNet", options=["-Q", "weka.classifiers.bayes.net.search.local.TAN"])
cls.build_classifier(data)

for index, inst in enumerate(data):
    pred = cls.classify_instance(inst)
    dist = cls.distribution_for_instance(inst)
    print(str(index+1) + ": label index=" + str(pred) + ", class distribution=" + str(dist))
```
